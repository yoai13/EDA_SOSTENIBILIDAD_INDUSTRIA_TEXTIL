# **HIPÓTESIS 1:** *Las marcas con una calificación de sostenibilidad más alta tienden a tener una menor huella de carbono, menor uso de agua y menor producción de residuos.*

```python
#Análisis por grupos de calificación de sostenibilidad
grouped_data = df1.groupby('Sustainability_Rating')[['Carbon_Footprint_MT', 'Water_Usage_Liters', 'Waste_Production_KG']].mean()
grouped_data
```

```python
#El resultado de la columna Water_Usage_Liters está en notación científica. 2.524030×1000000=2524030
```

```python
#Gráfica 1: Promedio de Huella de Carbono por Calificación de Sostenibilidad
plt.figure(figsize=(6, 3))
plt.bar(grouped_data.index, grouped_data['Carbon_Footprint_MT'], color="#0f110f")
plt.xlabel('Calificación de Sostenibilidad')
plt.ylabel('Promedio de Huella de Carbono (MT)' , fontsize=8)
plt.title('Promedio de Huella de Carbono por Calificación de Sostenibilidad')
plt.xticks(grouped_data.index)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.savefig("../images/promedio_carbon_sostenibilidad.jpg")

#Gráfica 2: Promedio de Uso de Agua por Calificación de Sostenibilidad
plt.figure(figsize=(6, 3))
plt.bar(grouped_data.index, grouped_data['Water_Usage_Liters'], color="#1db6d1")
plt.xlabel('Calificación de Sostenibilidad')
plt.ylabel('Promedio de Uso de Agua (Liters)' , fontsize=8)
plt.title('Promedio de Uso de Agua por Calificación de Sostenibilidad')
plt.xticks(grouped_data.index)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.savefig("../images/promedio_agua_sostenibilidad.jpg")

#Gráfica 3: Promedio de Producción de Desperdicios por Calificación de Sostenibilidad
plt.figure(figsize=(6, 3))
plt.bar(grouped_data.index, grouped_data['Waste_Production_KG'], color="#925C0B")
plt.xlabel('Calificación de Sostenibilidad')
plt.ylabel('Promedio de Producción de Desperdicios (KG)', fontsize=8)
plt.title('Promedio de Producción de Desperdicios por Calificación de Sostenibilidad')
plt.xticks(grouped_data.index)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.savefig("../images/promedio_desperdicios_sostenibilidad.jpg")

plt.show()
```

```python
#Cálculo de correlaciones: Calcula la correlación entre la calificación de sostenibilidad y las métricas de impacto ambiental. Una correlación negativa indicaría que a mayor calificación de sostenibilidad, menor es el impacto ambiental.
correlation_carbon = df1['Sustainability_Rating'].corr(df1['Carbon_Footprint_MT'])
print("Correlación entre la Calificación de Sostenibilidad y la Huella de Carbono", correlation_carbon)

correlation_water = df1['Sustainability_Rating'].corr(df1['Water_Usage_Liters'])
print("Correlación entre la Calificación de Sostenibilidad y el Uso de Agua", correlation_water)

correlation_waste = df1['Sustainability_Rating'].corr(df1['Waste_Production_KG'])
print("Correlación entre la Calificación de Sostenibilidad y la Producción de Desperdicios", correlation_waste)
```

```python
#Visualizo la correlación entre la calificación de sostenibilidad y las métricas de impacto ambiental
#Creo un DataFrame para facilitar la graficación
correlations_data = pd.DataFrame({
    'Métrica': ['Carbono', 'Agua', 'Desperdicios'],
    'Correlación': [correlation_carbon, correlation_water, correlation_waste]
})

#Gráfico de barras horizontales
plt.figure(figsize=(8, 4))
plt.barh(correlations_data['Métrica'], correlations_data['Correlación'], color=["#0f110f", "#1db6d1", "#925C0B"])
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Línea de referencia en cero
plt.ylabel('Métrica de Impacto Ambiental')
plt.xlabel('Correlación con Calificación de Sostenibilidad')
plt.title('Correlación entre la Calificación de Sostenibilidad y Métricas de Impacto Ambiental')

# Añadir los valores de correlación al lado de las barras
for index, value in enumerate(correlations_data['Correlación']):
    plt.text(value, index, f'{value:.4f}', va='center', ha='left', color= "red")

#Muestro el gráfico
plt.savefig("../images/correlacion_sostenibilidad_metricas")
plt.show()
```

```python
#Visualización Directa con Gráficos de Caja (Box Plots):
df1.dropna(subset=['Material_Type', 'Average_Price_USD'])

# Crear el gráfico de caja
plt.figure(figsize=(12, 8))
sns.boxplot(x='Material_Type', y='Average_Price_USD', data=df1)
plt.xlabel('Tipo de Material')
plt.ylabel('Precio Promedio (USD)')
plt.title('Distribución del Precio Promedio por Tipo de Material')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()
plt.show()
```

**Análisis del Gráfico de Caja para cada Tipo de Material:**

Para cada tipo de material, el gráfico de caja proporciona la siguiente información sobre la distribución del precio promedio de los productos hechos con ese material:

Caja (El rectángulo azul): Contiene el 50% central de los datos. Su borde inferior representa el primer cuartil (Q1, el 25% de los datos está por debajo de este valor) y su borde superior representa el tercer cuartil (Q3, el 75% de los datos está por debajo de este valor). La altura de la caja representa el rango intercuartílico (IQR = Q3 - Q1), que es una medida de la dispersión del 50% central de los datos.
Línea horizontal dentro de la caja: Representa la mediana (el valor central de los datos, el 50% de los precios están por debajo y el 50% por encima).
Bigotes (Las líneas verticales que se extienden desde la caja): Intentan cubrir la mayor parte del resto de la distribución de los datos, excluyendo los valores atípicos. Generalmente se extienden hasta 1.5 veces el IQR desde los cuartiles.
Puntos fuera de los bigotes (No hay puntos visibles en este gráfico): Representan valores atípicos (outliers), que son precios inusualmente altos o bajos en comparación con el resto de los datos para ese tipo de material. En este gráfico en particular, no se observan puntos fuera de los bigotes, lo que sugiere que no hay valores atípicos extremos evidentes para ninguno de los materiales representados.
Comparación entre los Tipos de Material:

Al comparar las cajas para los diferentes tipos de materiales, podemos observar lo siguiente sobre sus precios promedio:

Mediana: Las líneas dentro de las cajas parecen estar a niveles bastante similares para todos los tipos de materiales, sugiriendo que el precio promedio típico (la mediana) no varía drásticamente entre ellos.
Rango Intercuartílico (IQR): La altura de las cajas es también relativamente similar entre los diferentes materiales, lo que indica que la dispersión del 50% central de los precios es comparable.
Rango General (Longitud de los bigotes): La longitud de los bigotes también parece similar, sugiriendo que el rango general de precios (excluyendo posibles valores atípicos) es comparable entre los materiales.
Conclusión Preliminar:

A partir de este gráfico de caja, no se observa una diferencia drástica y evidente en la distribución del precio promedio entre los diferentes tipos de materiales representados (Tencel, Vegan Leather, Bamboo Fabric, Recycled Polyester, Hemp y Organic Cotton). ## **HIPÓTESIS 15**: Hipótesis de Correlación entre Impactos: "Existe una correlación entre diferentes métricas de impacto ambiental (por ejemplo, huella de carbono y uso de agua) a nivel de país." (Podemos calcular correlaciones agregadas por país).

```python
#Calculo el promedio de las métricas de impacto ambiental por país
promedio_impacto_por_pais = df1.groupby('Country')[['Carbon_Footprint_MT', 'Water_Usage_Liters', 'Waste_Production_KG']].mean()

print("Promedio de Métricas de Impacto Ambiental por País:\n", promedio_impacto_por_pais)
```

```python
#CalculO la matriz de correlación entre las métricas a nivel de país
matriz_correlacion_pais = promedio_impacto_por_pais.corr(method='pearson')

print("\nMatriz de Correlación entre Métricas de Impacto Ambiental a Nivel de País:\n", matriz_correlacion_pais)
```

```python
#Visualización de la matriz de correlación con un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion_pais, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre Métricas de Impacto Ambiental Promedio por País')
plt.tight_layout()
plt.show()
```

```python
#Analizo de correlación específica entre dos métricas
correlacion_carbono_agua = promedio_impacto_por_pais['Carbon_Footprint_MT'].corr(promedio_impacto_por_pais['Water_Usage_Liters'], method='pearson')
print(f"\nCorrelación entre Promedio de Huella de Carbono y Promedio de Uso de Agua por País: {correlacion_carbono_agua:.2f}")

correlacion_carbono_residuos = promedio_impacto_por_pais['Carbon_Footprint_MT'].corr(promedio_impacto_por_pais['Waste_Production_KG'], method='pearson')
print(f"Correlación entre Promedio de Huella de Carbono y Promedio de Producción de Residuos por País: {correlacion_carbono_residuos:.2f}")

correlacion_agua_residuos = promedio_impacto_por_pais['Water_Usage_Liters'].corr(promedio_impacto_por_pais['Waste_Production_KG'], method='pearson')
print(f"Correlación entre Promedio de Uso de Agua y Promedio de Producción de Residuos por País: {correlacion_agua_residuos:.2f}")
```

```python
#Gráfico de dispersión para visualizar la relación entre dos métricas
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Carbon_Footprint_MT', y='Water_Usage_Liters', data=promedio_impacto_por_pais)
for i, row in promedio_impacto_por_pais.iterrows():
    plt.annotate(i, (row['Carbon_Footprint_MT'], row['Water_Usage_Liters']), textcoords="offset points", xytext=(0,10), ha='center')
plt.title('Correlación entre Promedio de Huella de Carbono y Promedio de Uso de Agua por País')
plt.xlabel('Promedio de Huella de Carbono (MT)')
plt.ylabel('Promedio de Uso de Agua (Litros)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Gráfico de dispersión para visualizar la relación entre dos métricas
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Carbon_Footprint_MT', y='Waste_Production_KG', data=promedio_impacto_por_pais)
for i, row in promedio_impacto_por_pais.iterrows():
    plt.annotate(i, (row['Carbon_Footprint_MT'], row['Waste_Production_KG']), textcoords="offset points", xytext=(0,10), ha='center')
plt.title('Correlación entre Promedio de Huella de Carbono y Promedio de Producción de Residuos por País')
plt.xlabel('Promedio de Huella de Carbono (MT)')
plt.ylabel('Promedio de Producción de Residuos (Kg)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Este primer conjunto de resultados muestra el promedio de la Huella de Carbono (en toneladas métricas - MT), el Uso de Agua (en litros) y la Producción de Residuos (en kilogramos - KG) para las marcas dentro de cada país.

Huella de Carbono: Los promedios de huella de carbono parecen estar en un rango relativamente estrecho entre los países, desde aproximadamente 241 MT (Alemania) hasta 256 MT (China). China e India muestran los promedios ligeramente más altos, mientras que Alemania y Francia tienen los más bajos en este conjunto de países.
Uso de Agua: El promedio de uso de agua varía más significativamente entre países. El Reino Unido muestra el promedio más alto, seguido de Japón y Australia. Estados Unidos y Francia tienen los promedios de uso de agua más bajos. La escala de los valores (en millones de litros) es mucho mayor que la de la huella de carbono.
Producción de Residuos: Los promedios de producción de residuos también muestran variación, aunque en una escala menor en comparación con el uso de agua. China, Italia y Estados Unidos tienen promedios ligeramente más altos, mientras que Francia e India tienen los más bajos. La escala está en decenas de miles de kilogramos.
Análisis de la Matriz de Correlación entre Métricas de Impacto Ambiental a Nivel de País:

Esta matriz muestra las correlaciones de Pearson entre los promedios de las tres métricas de impacto ambiental a nivel de país.

Correlación entre Huella de Carbono y Uso de Agua: El valor de -0.32 indica una correlación negativa débil. Esto sugiere que, en promedio a nivel de país, puede haber una ligera tendencia a que los países con una mayor huella de carbono promedio tiendan a tener un menor uso de agua promedio, y viceversa. Sin embargo, la correlación es bastante débil, por lo que esta relación no es muy fuerte ni consistente.
Correlación entre Huella de Carbono y Producción de Residuos: El valor de 0.39 indica una correlación positiva moderada. Esto sugiere que, en promedio a nivel de país, existe una tendencia a que los países con una mayor huella de carbono promedio también tiendan a tener una mayor producción de residuos promedio. Esta relación es más fuerte que la anterior.
Correlación entre Uso de Agua y Producción de Residuos: El valor de 0.04 indica una correlación positiva muy débil, prácticamente nula. Esto sugiere que no hay una relación lineal significativa entre el promedio de uso de agua y el promedio de producción de residuos a nivel de país.

La relación entre la huella de carbono y el uso de agua es negativa y débil, lo que indica una ligera tendencia opuesta, pero no una regla fuerte.
